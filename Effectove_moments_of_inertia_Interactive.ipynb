```
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')
from ipywidgets import interactive

def effMIO(c):

    c_value = c * 1e6 # N/m2
    Lenghts = [0.8, 1, 1.2, 1.4, 1.6, 1.8 , 2 , 2.2, 2.4, 2.6, 2.8, 3, 4, 5, 6, 7 ] # meters

    massprop1 = '''
     ----------------   REGIONS   ----------------
    Area:                    239.8695
    Perimeter:               279.5645
    Bounding box:         X: 1216.9148  --  1283.9148
                          Y: 425.9931  --  443.2585
    Centroid:             X: 1256.1175
                          Y: 431.9514
    Moments of inertia:   X: 44763661.7830
                          Y: 378546729.2240
    Product of inertia:  XY: -130156506.8001
    Radii of gyration:    X: 431.9916
                          Y: 1256.2390

    '''

    massprop2 = '''
    ----------------   REGIONS   ----------------
    Area:                    226.8829
    Perimeter:               257.4631
    Bounding box:         X: 1243.9148  --  1283.9148
                          Y: 477.7276  --  502.9931
    Centroid:             X: 1262.4601
                          Y: 491.3802
    Moments of inertia:   X: 54801331.5538
                          Y: 361638813.7324
    Product of inertia:  XY: -140748752.6472
    Radii of gyration:    X: 491.4674
                          Y: 1262.5154
    '''

    import numpy as np
    L = np.array(Lenghts)

    massprop1parsed = massprop1.strip().split()
    A1 =  float(massprop1parsed[massprop1parsed.index('Area:')+1]) * 1e-6
    y1 =  float(massprop1parsed[massprop1parsed.index('Centroid:')+4]) * 1e-3
    Ix =  float(massprop1parsed[massprop1parsed.index('inertia:')+2]) * 1e-12
    Ix1 = Ix - A1*y1**2
#     print(A1, y1, Ix1)

    massprop2parsed = massprop2.strip().split()
    A2 =  float(massprop2parsed[massprop2parsed.index('Area:')+1]) * 1e-6
    y2 =  float(massprop2parsed[massprop2parsed.index('Centroid:')+4]) * 1e-3
    Ix =  float(massprop2parsed[massprop2parsed.index('inertia:')+2]) * 1e-12
    Ix2 = Ix - A2*y2**2
#     print(A2, y2, Ix2)



    def lam_param(c, a, l, E, Is, nu):
        return np.sqrt(c*a**2*l**2/E/Is*1/nu/(1-nu))

    def C_param(lam):
        return lam**2 / (np.pi**2 + lam**2)

    def nu_param(A1, a1, A2, a2, Is):
        return (A1*a1**2 + A2*a2**2) / Is

    def rigicMOI(I1, I2, A1, a1, A2, a2):
        return I1 + I2 + A1*a1**2 + A2*a2**2

    def effMOI(Is, nu, C):
        return Is * (1-nu)/(1-nu*C)

    E = 70 * 1e9
    a = np.absolute(y1-y2)
    a1 = A2*a/(A1+A2)
    a2 = A1*a/(A1+A2)
    # print(a1, a2)

    Is = rigicMOI(Ix1, Ix2, A1, a1, A2, a2)  
    nu = nu_param(A1, a1, A2, a2, Is)
    lam = lam_param(c_value, a, L, E, Is, nu)
    C = C_param(lam)
    Ief = effMOI(Is, nu, C)

#     print('Ieff [cm4]')
#     print(*Ief*1e8 , sep='\n') 

    plt.plot(L, Ief*1e8)
    plt.ylim(0, Is*1e8*1.1)
    
# effMIO(25)
interactive_plot_moi = interactive(effMIO, c=(10, 100))
display(interactive_plot_moi)
```
